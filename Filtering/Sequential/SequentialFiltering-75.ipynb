Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Open The Data

```python
df = pd.read_csv('../../train.csv')
df.head()
```

Take 75% of the data

```python
df_sample = df.sample(frac=0.75, random_state=42)
```

```python
rows = df_sample.to_dict(orient="records")
```

```python
rows
```

Data Filtering Sequential Loop

```python
filtered_rows = []
for row in rows:
    if row["trip_duration"] >= 1000:
        filtered_rows.append(row)

# Check how many rows passed the filter
print(f"Filtered rows: {len(filtered_rows)}")
```

```python
df_filtered = pd.DataFrame(filtered_rows)
df_filtered.head()
```
